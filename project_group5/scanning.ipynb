# Make scanning of runtime for different sizes of input in cpp

Obs!

Program input cannot be provided by a placeholder e.g. x=32, y=32

```python
## Scanning for order = 4
output = [];
itera = 4
size = []

for iteration in range(itera):
    u = !./stencil2d-pro.x+orig --nx 16 --ny 16 --nz 64 --num_iter 1024 --order 4
    uu0 = u[0].split(',')
    uuu = [float(uu) for uu in uu0]
    output = output + uuu
    size.append(16 * 16 * 64)
    size.append(16 * 16 * 64)
    
for iteration in range(itera):
    u0 = !./stencil2d-pro.x+orig --nx 32 --ny 32 --nz 64 --num_iter 1024 --order 4
    u00 = u0[0].split(',')
    u000 = [float(uu) for uu in u00]
    output = output + u000
    size.append(32 * 32 * 64)
    size.append(32 * 32 * 64)

for iteration in range(itera):
    u1 = !./stencil2d-pro.x+orig --nx 32 --ny 64 --nz 64 --num_iter 1024 --order 4
    u11 = u1[0].split(',')
    u111 = [float(uu) for uu in u11]
    output = output + u111
    size.append(32 * 64 * 64)
    size.append(32 * 64 * 64)
    
for iteration in range(itera):
    u2 = !./stencil2d-pro.x+orig --nx 64 --ny 64 --nz 64 --num_iter 1024 --order 4
    u22 = u2[0].split(',')
    u222 = [float(uu) for uu in u22]
    output = output + u222
    size.append(64 * 64 * 64)
    size.append(64 * 64 * 64)
    
for iteration in range(itera):
    u3 = !./stencil2d-pro.x+orig --nx 64 --ny 128 --nz 64 --num_iter 1024 --order 4
    u33 = u3[0].split(',')
    u333 = [float(uu) for uu in u33]
    output = output + u333
    size.append(64 * 128 * 64)
    size.append(64 * 128 * 64)
    
for iteration in range(itera):
    u4 = !./stencil2d-pro.x+orig --nx 128 --ny 128 --nz 64 --num_iter 1024 --order 4
    u44 = u4[0].split(',')
    u444 = [float(uu) for uu in u44]
    output = output + u444
    size.append(128 * 128 * 64)
    size.append(128 * 128 * 64)
    
for iteration in range(itera):
    u5 = !./stencil2d-pro.x+orig --nx 128 --ny 192 --nz 64 --num_iter 1024 --order 4
    u55 = u5[0].split(',')
    u555 = [float(uu) for uu in u55]
    output = output + u555
    size.append(128 * 192 * 64)
    size.append(128 * 192 * 64)
    
for iteration in range(itera):
    u6 = !./stencil2d-pro.x+orig --nx 192 --ny 192 --nz 64 --num_iter 1024 --order 4
    u66 = u6[0].split(',')
    u666 = [float(uu) for uu in u66]
    output = output + u666
    size.append(192 * 192 * 64)
    size.append(192 * 192 * 64)

np.save('output4.npy',output) 
np.save('size.npy',size) 
```
