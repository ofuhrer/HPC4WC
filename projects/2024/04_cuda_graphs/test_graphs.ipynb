```python
import numpy as np
import matplotlib.pyplot as plt
import os

data_cuda = np.loadtxt('CUDA_results_128_it.txt', delimiter=',', skiprows=1)
# 0=nx, 1=ny, 2=nz, 3=num_iter, 4=time(seconds), 5=data_size(bytes)
data_normal = np.loadtxt('normal_results_128_it.txt', delimiter=',', skiprows=1)

runtime_per_gridpoint_cuda = data_cuda[:,4]/(data_cuda[:,0] * data_cuda[:,1] * data_cuda[:,2])
runtime_per_gridpoint_normal = data_normal[:,4]/(data_normal[:,0] * data_normal[:,1] * data_normal[:,2])

fig, ax = plt.subplots(figsize=(7, 5))
ax.loglog((data_cuda[:,5]/1024)/1024, runtime_per_gridpoint_cuda[:] * 10**6, ".", label="CUDA") 
ax.loglog((data_normal[:,5]/1024)/1024, runtime_per_gridpoint_normal[:] * 10**6, ".", label="normal") 
ax.grid(which="both")
ax.set_xlabel("Working set size [MiB]")
ax.set_ylabel("Runtime per gridpoint [μs]")
ax.legend()
fig.show()
```
